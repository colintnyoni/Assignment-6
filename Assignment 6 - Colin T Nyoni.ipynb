```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Q1
# load built-in gapminder dataset from plotly 
gapminder = px.data.gapminder() 
gapminder
```

```python
#Q1
# Number of entries in each continent
gapminder["continent"].value_counts()
```

```python
#Q1
#Bar Graph representation of all contries in each continent
gapminder["continent"].value_counts().plot(kind='bar')
```

```python
#Q1
#List of all data for Asian countries
gapminder_Asia = gapminder[gapminder["continent"]== "Asia"]
gapminder_Asia
```

```python
#Q1
#Data frame of all Asian countries and their mean life expectancy
gapminder_Asia_mean_lifeExp=gapminder_Asia.groupby(['country','continent','iso_alpha'])['lifeExp'].mean().reset_index()\
        .rename(columns={"lifeExp":"mean_lifeExp"})
gapminder_Asia_mean_lifeExp
```

```python
#Q1: Geomap Represenation of Asian country's mean life expectancy
fig = px.choropleth(gapminder_Asia_mean_lifeExp, locations='iso_alpha', color='mean_lifeExp', hover_name='country', 
                     color_continuous_scale=px.colors.sequential.Plasma, projection='natural earth')
fig.show() 
```

```python
#Q2
#List of all data from countries in South America nd Europe
gapminder_Europe_Americas=gapminder[(gapminder["continent"]== "Europe") | (gapminder["continent"]== "Americas")]
gapminder_Europe_Americas
```

```python
#Q2
#Standard Deviation of GDP for each country in South America and Europe
gapminder_Europe_Americas__std_gdp=gapminder_Europe_Americas.groupby(['country','continent','iso_alpha'])['gdpPercap'].std().reset_index()\
        .rename(columns={"gdpPercap":"standard_deviation_gdpPercap"})
gapminder_Europe_Americas__std_gdp
```

```python
#Geographic representation of the Standard Deviation of the GDPs of each country in South America and Europe
fig = px.choropleth(gapminder_Europe_Americas__std_gdp, locations='iso_alpha', color='standard_deviation_gdpPercap', hover_name='country', 
                     color_continuous_scale=px.colors.sequential.Plasma, projection='natural earth')
fig.show() 
```

```python
#Q2
#Bar Graph Representation of Standard Deviations of the GDP of each country in South America and Europe
#let's add color by lifeExp and other parameters 
fig = px.bar(gapminder_Europe_Americas__std_gdp, x='country', y='standard_deviation_gdpPercap', color='continent', labels={'gdpPercap': 'standard dev of gdpPercap ',}, height=400,)
fig.show()
```

```python
#Q3
#minimum population in each country over the 3 dacades
gapminder_min_pop=gapminder.groupby(['country','continent','iso_alpha'])['pop'].min().reset_index()\
        .rename(columns={"pop":"population"})
gapminder_min_pop
```

```python
#Q3
#maximum population in each country over the 3 dacades
gapminder_max_pop=gapminder.groupby(['country','continent','iso_alpha'])['pop'].max().reset_index()\
        .rename(columns={"pop":"population"})
gapminder_max_pop
```

```python
#Q3 Growth in population of all the countries over the 3 decades
gapminder_max_pop['population_growth']=gapminder_max_pop.population.subtract (gapminder_min_pop.population)
gapminder_max_pop
```

```python
#Q3
#Growth in population over the 3 dacades
fig = px.choropleth(gapminder_max_pop, locations='iso_alpha', color='population_growth', hover_name='country', 
                     color_continuous_scale=px.colors.sequential.Plasma, projection='natural earth')
fig.show() 
```
